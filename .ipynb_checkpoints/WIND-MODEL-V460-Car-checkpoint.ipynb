# Extended atmospheres model program

This script computes the extended atmosphere for the case of V460 Car, but it can be used for different targets. You would need: the initial MARCS model (so Teff, logg, metallicity) and you need to know the luminosity and mass of the star approx in order to transform the original MARCS model to the right parameters.

To transform the MARCS model, we use the conversion factors found in "/model/Read_MIST_tracks-rescale-MARCS-factor.ipynb" programs. Essentially, after talking with Betrand we found out that we cannot just assume a MARCS model and go with it, because the luminosities and radius would be different for each target. We have to then convert the MARCS using a factor found with the MIST evolutionary tracks depending on the metallicity, radius and mass we have of the star. 

Afterwards, we compute the temperature, density, opacity profile and run it with Turbospectrum in alva format. Then, we open it again to check if the dlogtau is really between 0.05 and 0.08, and interpolate if it is not the case. With the new model, we re-run Turbospectrum and now it should not crash.

```python
#From the conversion factor from "Read_MIST_tracks-rescale-MARCS-factor.ipynb" 
#For a Model with these parameters:
#4.405595598788188 3636.6409926665765 0.2640370906361084 (luminosity, teff, logg)
#10.815312109194467 401.8593633679475 (mass, radius)

#the luminosity corresponds to factor2 in "Read_MIST_tracks-rescale-MARCS-factor.ipynb"
factors=[1.2653270749375707]
mass_losses=[-7, -6.5,-6, -5, -4, -5.50, -3.50, -4.5, -3.75] 
```

```python

#luminosity of the re-scaled MARCS model from the MIST data
logL_rsg=[4.4055956]
```

```python
#The initial MARCS model we use
models=["/model/s3600_g-0.5_m15._t05_st_z+0.00_a+0.00_c+0.00_n+0.00_o+0.00_r+0.00_s+0.00.mod"
        
    ]
```

```python
#assign a luminosity and mass_loss to the model
u=0
j=1
#10**5.3/10**5
lum="%.1f"%logL_rsg[u]
factor=factors[u]
mass_lossesi=mass_losses[j]
#mass_lossesi=-6.5
mass_lossesi
```

```python
#READ MARCS MODEL 
i=0
#model="/home/ariggonz/Turbospectrum2019_MYWORK/COM-v19.1/models/s3600_g-0.5_m15._t05_st_z+0.00_a+0.00_c+0.00_n+0.00_o+0.00_r+0.00_s+0.00.mod"
data=pd.read_table("%s"%models[i], skiprows=24, nrows=56, sep="\s+", engine="python")
teff=data["T"]
depth=data["Depth"]
t_ross=data["lgTauR"]
datad=pd.read_table("%s"%models[i], skiprows=81, nrows=56, sep="\s+", engine="python")
density=datad["Density"]
kappar=datad["KappaRoss"]
constants=pd.read_table("%s"%models[i],skiprows=1, header=None, nrows=8, sep="\s+", engine="python")
mu=datad["Mu"]
data
```

```python
#Define the constants
vinf=25. #velocities in km/s
v0=0.1
vcore=3e-5
heff=8.6
beta=1.0
rmax=30. #in solar radius
rstar=constants[0][6]*np.sqrt(factor) #in cm here we re-scale the initial MARCS radius using this factor from MIST!
rsun=6.957e+10 #radius sun in cm 
ND=60 #n depth points 
Mdot_fid=3e-6 # Harper's Mdot estimate for a Ori

#Define parameters that can change (tau, Mdot, etc)
Mdot=10**mass_lossesi #mass loss in solar mass units
dlogmdot=np.log10(Mdot)
dlogtau=0.05
dlogtaumin=-8.



MMP=1.2 #mean mass/particle (set in MARCS model?)
kb_cgs=1.38064852e-16 #boltzmann constant in cgs
msun_cgs=1.989e+33#in cgs
vinf_cgs=vinf*1e5
sec_per_yr=3600.*24.*365.24
print(rstar/rsun)
print(10*factor)
print(dlogmdot)
```

```python
#Read Harper's model
datah=pd.read_table("/model/harper_data.dat",sep=",",engine="python")
rstar_harp=datah["R/R"]
print(rstar_harp)
temp_harp=datah["T"]
lognh_harp=datah["logn_h"]
logne_harp=datah["logn_e"]
logNcolH_harp=datah["logNH"]
rstar_harp *= rstar #(rstar_harp=rstar_harp*rstar) IN CM#
#rstar_harp*rstar
```

```python
radius=rstar-depth
```

```python
#New rgrid based on harper
nharp=len(rstar_harp[7:43]) #we are extending the model using the harper radius points
nr=len(radius)
ndp_marcs=len(radius)
new_rgrid=np.zeros(nharp+nr) #new radius grid with initial marcs + extra Harper radius points to extend the atmosphere
new_rgrid[:nr]=np.flip(radius) #this is the initial MARCS radius points
new_rgrid[nr:]=rstar_harp[7:43] #then we extend the model to the harper

#plot the new radius grid
#fig=plt.figure(figsize=(10,10)) 
#ax=fig.add_subplot((111))
#ax.plot(new_rgrid/rstar-0.99, range(len(new_rgrid)), ".")
#rstar_harp[7:43]/rstar
```

```python
#We have the radius, now we need the temperature with Harper
new_Tgrid=np.zeros(nharp+nr) #define new temperature with MARCS points + Harper extension
new_Tgrid[:nr]=np.flip(teff) #MARCS temperature
new_Tgrid[nr:]=temp_harp[7:43] #Harper temperature

#The region between Marcs and Harper (~1 R*) doesn't have many points, we interpolate to obtain more in that region
interpfunc2=interpolate.interp1d((new_rgrid[nr-1:nr+1]),new_Tgrid[nr-1:nr+1],kind="linear",fill_value="extrapolate")
#these are the radius where we interpolate the temperature
add_rgrid2=[1.06*rstar,1.07*rstar]
Tgrid2=interpfunc2(add_rgrid2) #new interpolated temperature
#new_rgrid[nr-1:nr+1]/rstar 
```

```python
#Now we write it all in a new variable to make it clearer
#radius with: MARCS points, Harper points, the interpolated points
nharp=len(rstar_harp[7:43])#[13:39]
nr=len(radius)
ndp_marcs=len(radius)
new_rgrid=np.zeros(nharp+nr+len(add_rgrid2))
new_rgrid[:nr]=np.flip(radius)
new_rgrid[nr:nr+len(add_rgrid2)]=add_rgrid2
new_rgrid[nr+len(add_rgrid2):]=rstar_harp[7:43]
#print(new_rgrid/rstar)

#Temperature with: Marcs points, Harper points, interpolated points
new_Tgrid=np.zeros(nharp+nr+len(add_rgrid2))
new_Tgrid[:nr]=np.flip(teff)
new_Tgrid[nr:nr+len(add_rgrid2)]=Tgrid2
new_Tgrid[nr+len(add_rgrid2):]=temp_harp[7:43]


#We plot the new Harper temperature profile
fig= plt.figure(figsize=(10,5))
ax=fig.add_subplot((111))
ax.plot((new_rgrid/rstar),new_Tgrid,".", c="b", label="Harper")
#ax.plot((rgrid/rstar)-0.99, Tgrid_rad,".", c="r", label="rad")
ax.set_xscale("log")
ax.set_xlim(0,10)
ax.set_xlabel(r"$R/R_{*}-0.99$")
ax.set_ylabel(r"$T_{eff}$ (K)")
ax.legend()
```

```python
#Here we see the points on our R/R* grid
new_rgrid/rstar
```

```python
#This is the temperature for Harper that we have just found out
Tgrid=new_Tgrid

#Now we get the radiative equilibrium Tgrid 
Tgrid_rad=np.zeros(nr+nharp+len(add_rgrid2)) #same length as Harper
Tgrid_rad[:nr]=np.flip(teff) #MARCs for the non-extension
Tgrid_rad[nr:]=Tgrid_rad[nr-1]*np.sqrt(new_rgrid[nr-1]/new_rgrid[nr:]) #rad equilibrium definition

#Plot
fig= plt.figure(figsize=(10,5))
ax=fig.add_subplot((111))
ax.plot((new_rgrid/rstar)-0.99,Tgrid,".", c="b", label="Harper")
ax.plot((new_rgrid/rstar)-0.99, Tgrid_rad,".", c="r", label="rad")
ax.set_xscale("log")
ax.set_xlim(0,10)
ax.set_xlabel(r"$R/R_{*}-0.96$")
ax.set_ylabel(r"$T_{eff}$ (K)")
ax.legend()

#print(rgrid/rstar)
print(Tgrid[0:nr],Tgrid_rad[0:nr])
print((new_rgrid/rstar)-0.99)
#fig.savefig("temperature_profile.png")
```

```python
#Harpers analytical density 
#Define different gamma and beta parameters
gamma_harp1=0.45
beta_harp1=-1.10

#Get the density profile
dens_ext=min(density)
zz=new_rgrid/rstar
dens_wind1=(dens_ext/zz**2)*(1.-(0.998/zz)**gamma_harp1)**beta_harp1 #beta profile for dens
mdot_curr1=4.*np.pi*(new_rgrid[-1])**2*dens_wind1[-1]*(vinf_cgs)/msun_cgs*sec_per_yr 
#print(dens_ext)
dens_wind1 *=(Mdot/mdot_curr1)
#print(dens_wind)
dens_wind1[np.where(np.isinf(dens_wind1))]=0.0
dens_wind1[np.where((zz-0.99)<0.01)]=0.0
#print(dens_wind)

#extrapolate from outer star
upper_star_rho=np.flip(density)[-3:]
upper_star_rad=np.flip(radius)[-3:]
fp=np.polyfit(np.log10(upper_star_rad),np.log10(upper_star_rho),1)
p=np.poly1d(fp)
dens_inner1=10**p(np.log10(new_rgrid))
#print(dens_inner)
dens_inner1[:nr]=np.flip(density)

#total density profile
dens_total1=dens_wind1+dens_inner1
msun=1.989e+30
#inferred velocity profile
vel=1./(dens_total1/(Mdot*msun_cgs/31557600.)*(4*np.pi*(new_rgrid)**2))/1e5

#print(dens_wind)
```

```python
#This is the equivalent in python to the IDL tabulate function. In the end I don't use it though...
def idl_tabulate(x, f, p=5) :
    def newton_cotes(x, f) :
        if x.shape[0] < 2 :
            return 0
        rn = (x.shape[0] - 1) * (x - x[0]) / (x[-1] - x[0])
        weights = integrate.newton_cotes(rn)[0]
        return (x[-1] - x[0]) / (x.shape[0] - 1) * np.dot(weights, f)
    ret = 0
    for idx in range(0, x.shape[0], p - 1) :
        ret += newton_cotes(x[idx:idx + p], f[idx:idx + p])
    return ret
```
