# Flux calibration for MATISSE data
This is an example for the target V602 Car, for the rest of the targets the script is the same changing the input 
files and MARCS models.

IMPORTANT: change the url of the files location (now it is set to my personal computer, change to where your files are located.

```python
#import necessary python modules
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.utils.data import get_pkg_data_filename
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
#open RAW CALIBRATOR
image_filec = get_pkg_data_filename('/home/ariggonz/Desktop/MATISSE-data-model-comp/V602-Car/CAL_L_V602-Car_HD94683_CALIB_RAWT_2023-01-20T07_12_27.2223_1.fits')

image_wavelength=fits.getdata(image_filec, ext=3)
image_flux=fits.getdata(image_filec, ext=8)
image_vis2=fits.getdata(image_filec, ext=4)
```

```python
#check header info
hdul = fits.open(image_filec)
hdul[0].header
```

```python
#get the cal wavelength on a 1D array
wc=[]
for i in range(len(image_wavelength[:])):
    wc.append(image_wavelength[i][0])
#wc=np.flip(wc)
```

```python
# get the cal vis2 on a 2D array (accounting for of the 6 baselines)
vis2c=[[0 for x in range(len(image_vis2[0][4]))] for y in range(6)]
for i in range(6):
    for j in range(len(image_vis2[0][4])):
        #print(j)
        vis2c[i][j]=image_vis2[i][4][j]
```

```python
#there are 4 fluxes, do it one by one! not that efficient as an array though
fluxc0=[]
for i in range(len(image_flux[0][4])):
    fluxc0.append(image_flux[0][4][i])
    
fluxc1=[]
for i in range(len(image_flux[1][4])):
    fluxc1.append(image_flux[1][4][i])
    
fluxc2=[]
for i in range(len(image_flux[2][4])):
    fluxc2.append(image_flux[2][4][i])
    
fluxc3=[]
for i in range(len(image_flux[3][4])):
    fluxc3.append(image_flux[3][4][i])
```

```python
#plot the fluxes
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))
ax.plot(np.array(wc)*1e+6, fluxc0)
ax.plot(np.array(wc)*1e+6, fluxc1)
ax.plot(np.array(wc)*1e+6, fluxc2)
ax.plot(np.array(wc)*1e+6, fluxc3)
#ax.set_ylim(0, 1.1)
ax.set_title("CAL L reduced")
ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"Flux")
#fig.savefig("cal-l-reduced-1.7.6.png")
```

```python
#open MARCS model
marcs_flux=pd.read_table("/home/ariggonz/Desktop/MATISSE-data-model-comp/MARCS-models/s4000_g+0.0_m15._t05_st_z+0.00_a+0.00_c+0.00_n+0.00_o+0.00_r+0.00_s+0.00.flx", header=None)
m_flux=np.flip(marcs_flux[0])
marcs_w=pd.read_table("/home/ariggonz/Desktop/MATISSE-data-model-comp/MARCS-models/flx_wavelengths.vac", header=None)
m_w=np.flip(marcs_w[0])
```

```python
#Reduce the MARCS model to the MATISSE resolution
#get the delta lambda of the data
all_step=[]
for i in range(len(wc)-1):
    #print((wc[i]-wc[i+1])*1e+10) #step 
    all_step.append((wc[i]-wc[i+1])*1e+10)
```

```python
#get the MARCS models to same resolution for each data flux
x0=np.linspace(min(wnewc0)*1e+10, max(wnewc0)*1e+10, len(wnewc0)+1, endpoint=True)
len(x0)

cond0=(np.array(m_w)*1e-4<max(wnewc0)*1e+6)&(np.array(m_w)*1e-4>min(wnewc0)*1e+6)
cond0

digitized = np.digitize(m_w[cond0], x0)
bin_means_w0 = np.flip([m_w[cond0][digitized == i].mean() for i in range(1, len(x0))])
bin_means_w0

bin_means_f0 = np.flip([m_flux[cond0][digitized == i].mean() for i in range(1, len(x0))])
len(bin_means_f0)
```

```python
x1=np.linspace(min(wnewc1)*1e+10, max(wnewc1)*1e+10, len(wnewc1)+1, endpoint=True)
len(x1)

cond1=(np.array(m_w)*1e-4<max(wnewc1)*1e+6)&(np.array(m_w)*1e-4>min(wnewc1)*1e+6)
cond1

digitized = np.digitize(m_w[cond1], x1)
bin_means_w1 = np.flip([m_w[cond1][digitized == i].mean() for i in range(1, len(x1))])
bin_means_w1

bin_means_f1 = np.flip([m_flux[cond1][digitized == i].mean() for i in range(1, len(x1))])
len(bin_means_f1)
```

```python
x2=np.linspace(min(wnewc2)*1e+10, max(wnewc2)*1e+10, len(wnewc2)+1, endpoint=True)
len(x2)

cond2=(np.array(m_w)*1e-4<max(wnewc2)*1e+6)&(np.array(m_w)*1e-4>min(wnewc2)*1e+6)
cond2

digitized = np.digitize(m_w[cond2], x2)
bin_means_w2 = np.flip([m_w[cond2][digitized == i].mean() for i in range(1, len(x2))])
bin_means_w2

bin_means_f2 = np.flip([m_flux[cond2][digitized == i].mean() for i in range(1, len(x2))])
len(bin_means_f2)
```

```python
x3=np.linspace(min(wnewc3)*1e+10, max(wnewc3)*1e+10, len(wnewc3)+1, endpoint=True)
len(x3)

cond3=(np.array(m_w)*1e-4<max(wnewc3)*1e+6)&(np.array(m_w)*1e-4>min(wnewc3)*1e+6)
cond3

digitized = np.digitize(m_w[cond3], x3)
bin_means_w3 = np.flip([m_w[cond3][digitized == i].mean() for i in range(1, len(x3))])
bin_means_w3

bin_means_f3 = np.flip([m_flux[cond3][digitized == i].mean() for i in range(1, len(x3))])
len(bin_means_f3)
```

```python
#calibrate the fluxes with the flux of the SCI, the flux of the calibrator and the MARCS flux binned
#however, we need to normalize it so that the flux units make sense!! 
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))

ax.plot(np.array(bin_means_w2)*1e-4, np.array(flux2)/(np.array(fluxc2)/np.array(bin_means_f2)), label="flux 2")
ax.plot(np.array(bin_means_w1)*1e-4, np.array(flux1)/(np.array(fluxc1)/np.array(bin_means_f1)), label="flux 1")
ax.plot(np.array(bin_means_w3)*1e-4, np.array(flux3)/(np.array(fluxc3)/np.array(bin_means_f3)), label="flux 3")
ax.plot(np.array(bin_means_w0)*1e-4, np.array(flux0)/(np.array(fluxc0)/np.array(bin_means_f0)), label="flux 0")

ax.set_ylim(0,0.4e+6)

ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"Flux")
ax.legend()
```

```python
#normalize the fluxes
norm_marcs0=np.array(bin_means_f0)/np.mean(bin_means_f0)
norm_marcs1=np.array(bin_means_f1)/np.mean(bin_means_f1)
norm_marcs2=np.array(bin_means_f2)/np.mean(bin_means_f2)
norm_marcs3=np.array(bin_means_f3)/np.mean(bin_means_f3)
```

```python
#normalize the fluxes
norm_flux0=np.array(flux0)/np.mean(flux0)
norm_flux1=np.array(flux1)/np.mean(flux1)
norm_flux2=np.array(flux2)/np.mean(flux2)
norm_flux3=np.array(flux3)/np.mean(flux3)
```

```python
#normalize the fluxes
norm_fluxc0=np.array(fluxc0)/np.mean(fluxc0)
norm_fluxc1=np.array(fluxc1)/np.mean(fluxc1)
norm_fluxc2=np.array(fluxc2)/np.mean(fluxc2)
norm_fluxc3=np.array(fluxc3)/np.mean(fluxc3)
```

```python

#Calibrate the SCI with normalized fluxes

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))

ax.plot(np.array(bin_means_w2)*1e-4, (np.array(norm_flux2)/np.array(norm_fluxc2))*np.array(norm_marcs2), label="flux 2")
ax.plot(np.array(bin_means_w1)*1e-4, (np.array(norm_flux1)/np.array(norm_fluxc1))*np.array(norm_marcs1), label="flux 1")
ax.plot(np.array(bin_means_w3)*1e-4, (np.array(norm_flux3)/np.array(norm_fluxc3))*np.array(norm_marcs3), label="flux 3")
ax.plot(np.array(bin_means_w0)*1e-4, (np.array(norm_flux0)/np.array(norm_fluxc0))*np.array(norm_marcs0), label="flux 0")
#ax.set_xlim(3.88, 4.1)
ax.set_ylim(0, 2.5)
#ax.set_title("CAL L reduced")
ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"Flux")#$|V|^{2}$
ax.legend()
fig.savefig("flux-MATISSE-V602-Car.png") #save file
```

```python
#rename the normalized and calibrated SCI fluxes to one variable to make it less messy
mean_f0=(np.array(norm_flux0)/np.array(norm_fluxc0))*np.array(norm_marcs0)
mean_f1=(np.array(norm_flux1)/np.array(norm_fluxc1))*np.array(norm_marcs1)
mean_f2=(np.array(norm_flux2)/np.array(norm_fluxc2))*np.array(norm_marcs2)
mean_f3=(np.array(norm_flux3)/np.array(norm_fluxc3))*np.array(norm_marcs3)
```

```python
#do the mean of all the normalized fluxes
mean_flux=[]
for i in range(len(norm_flux2)):
    
    mean_flux_e=np.mean([mean_f0[i],
                  mean_f1[i],
                  mean_f2[i],
                  mean_f3[i]])
    mean_flux.append(mean_flux_e)
```

```python
#Plot the final, calibrated, normalized and mean FLUX of the SCI

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))

ax.plot(np.array(bin_means_w2)*1e-4, mean_flux, label="mean")

ax.set_ylim(0, 2.5)
ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"Flux")
fig.savefig("flux-MATISSE-V602-Car-mean.png") #save the figure
```
