```python
image_data = fits.getdata(image_file, ext=0)
```

```python
image_wavelength=fits.getdata(image_file, ext=3)
image_vis2=fits.getdata(image_file, ext=4)
image_array=fits.getdata(image_file, ext=2)
image_target=fits.getdata(image_file, ext=1)
```

```python
image_wavelength
w=[]
for i in range(len(image_wavelength[:])):
    w.append(image_wavelength[i][0])
```

```python
vis2=[[0 for x in range(len(image_vis2[0][4]))] for y in range(6)]
for i in range(6):
    for j in range(len(image_vis2[0][4])):
        #print(j)
        vis2[i][j]=image_vis2[i][4][j]
```

```python
u_coord=[]
for i in range(len(image_vis2)):
    u_coord.append(image_vis2[i][6])
v_coord=[]
for i in range(len(image_vis2)):
    v_coord.append(image_vis2[i][7])
baseline=np.sqrt(np.array(u_coord)**2+np.array(v_coord)**2)
baseline
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot((111))
ax.plot(np.array(w)*1e+6, vis2[-1], alpha=0.5, label="B=%.2f m"%baseline[-1])
for i in range(len(vis2)-1):
    ax.plot(np.array(w)*1e+6, vis2[i], label="B=%.2f m"%baseline[i])
ax.set_ylim(0, 0.2)
ax.set_xlim(8, 11.75)
#ax.set_title("KW Sgr calibrated")
ax.set_xlabel(r"wavelength $(\mu m)$", fontsize=15)
ax.set_ylabel(r"$|V|^{2}$", fontsize=15)
ax.legend(loc=2,prop={'size':15})
ax.tick_params(labelsize=15)
#fig.savefig("v602car-calibrated-n.png")
```
