# Plot MATISSE data
This script opens the fits file of MATISSE data and plots it. This is an example for V602 Car, but the script can be used for any file (just change the url of the file)


```python
#import necessary python packages
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.utils.data import get_pkg_data_filename
import numpy as np
%matplotlib notebook
```

```python
#here is where your file is located
files=['/home/ariggonz/Desktop/MATISSE-data-model-comp/V602-Car/SCI_V602-Car_TARGET_CAL_INT_2023-01-20T07_40_26.9617_4.fits']
```

```python
#Check if this is the BCD combined file. It should print "True"
for item in files:
    image_file_item = get_pkg_data_filename('%s'%item)
    #fits.info(image_file_item)
    hdul = fits.open(image_file_item)
    hdul[0].header
    print(hdul[0].header["HIERARCH ESO DRS BCD COMBINED"])
```

```python
#get the information of the .fits file
i=0
image_file = get_pkg_data_filename('%s'%files[i])
fits.info(image_file)
```

```python
#Get the information from each extension
image_data = fits.getdata(image_file, ext=0)
image_wavelength=fits.getdata(image_file, ext=3)
image_vis2=fits.getdata(image_file, ext=4)
image_vis=fits.getdata(image_file, ext=6)
image_flux=fits.getdata(image_file, ext=7)
image_array=fits.getdata(image_file, ext=2)
image_target=fits.getdata(image_file, ext=1)
```

```python
#transform the extracted wavelength into a 1D array (list)
w=[]
for i in range(len(image_wavelength[:])):
    w.append(image_wavelength[i][0])
```

```python
#transform the visibility squared extracted into a 2D array (for each wavelength and baseline)
vis2=[[0 for x in range(len(image_vis2[0][4]))] for y in range(6)]
for i in range(6):
    for j in range(len(image_vis2[0][4])):
        vis2[i][j]=image_vis2[i][4][j]
```

```python
#Same as v2 but for visibility
vis=[[0 for x in range(len(image_vis[0][4]))] for y in range(6)]
for i in range(6):
    for j in range(len(image_vis[0][4])):
        #print(j)
        vis[i][j]=image_vis[i][4][j]
```

```python
#Same as b2 but for the flux, this time it has no baselines so the 2nd dimension is 1 (list)
flux=[[0 for x in range(len(image_flux[0][4]))] for y in range(1)]
for i in range(1):
    for j in range(len(image_flux[0][4])):
        #print(j)
        flux[i][j]=image_flux[i][4][j]
```

```python
#Get the baselines 
baselines=[]
for i in range(0,6):
    b=np.sqrt(image_vis2[i][6]**2+image_vis2[i][7]**2)
    print(b)
    baselines.append(b)
```

```python
#Plot the visibility squared for each baseline
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))
for i in range(len(vis2)):
    ax.plot(np.array(w)*1e+6, vis2[i], label="B=%.2f m"%baselines[i])
ax.set_ylim(0, 1.3)
ax.set_title("V602 Car calibrated", fontsize=15)
ax.set_xlabel(r"wavelength $(\mu m)$", fontsize=15)
ax.set_ylabel(r"$|V|^{2}$", fontsize=15)
ax.legend(loc=1,prop={'size':15})
ax.tick_params(labelsize=15)
#fig.savefig("v602-car-calibrated.png") #save image
```

```python
#Plot the visibility for each baseline
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))
for i in range(len(vis2)):
    ax.plot(np.array(w)*1e+6, vis[i])
ax.set_ylim(0, 1.1)
ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"$V$")
#fig.savefig("hd-reduced-5.png")
```

```python
#Plot the uncalibrated flux
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot((111))
for i in range(len(flux)):
    ax.plot(np.array(w)*1e+6, flux[i])
ax.set_xlabel(r"wavelength $(\mu m)$")
ax.set_ylabel(r"$|V|^{2}$")
#fig.savefig("hd-reduced-5.png")
```
