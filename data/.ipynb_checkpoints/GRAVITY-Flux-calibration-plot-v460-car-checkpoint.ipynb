```python
#This function is taken from Kateryna Kravchenko's script, it reads the GRAVITY fits files

def read_FITS(input_file):

# add MJD    
   
# Open FITS file
    ff=pyfits.open('%s'%(input_file))

    OB = ff[2].data[0][1] # target name
    print(OB)
    
    wave_SC_P1 = [] # wavelength array for science spectrum and polarization 1
    bandwidth_SC_P1 = []
    for i in range(len(ff[3].data)):
        wave_SC_P1.append(ff[3].data[i][0]*1.e10)
        bandwidth_SC_P1.append(ff[3].data[i][1]*1.e10)
#    print wave_SC_P1,bandwidth_SC_P1  
    
    wave_SC_P2 = [] # wavelength array for science spectrum and polarization 2
    bandwidth_SC_P2 = []
    for i in range(len(ff[4].data)):
        wave_SC_P2.append(ff[4].data[i][0]*1.e10)
        bandwidth_SC_P2.append(ff[4].data[i][1]*1.e10)
#    print wave_SC_P2,bandwidth_SC_P2      
    
    wave_FT_P1 = [] # wavelength array for fringe tracking spectrum and polarization 1
    bandwidth_FT_P1 = []
    for i in range(len(ff[5].data)):
        wave_FT_P1.append(ff[5].data[i][0]*1.e10)
        bandwidth_FT_P1.append(ff[5].data[i][1]*1.e10)
#    print wave_FT_P1,bandwidth_FT_P1  
    
    wave_FT_P2 = [] # wavelength array for fringe tracking spectrum and polarization 2
    bandwidth_FT_P2 = []
    for i in range(len(ff[6].data)):
        wave_FT_P2.append(ff[6].data[i][0]*1.e10)
        bandwidth_FT_P2.append(ff[6].data[i][1]*1.e10)
#    print wave_FT_P2,bandwidth_FT_P2  
    
# squared visibility for the FT and SC -> for P1 and P2 at 6 baselines    
# each of variables is a list with 6-baseline datasets: each dataset is an array with 
# length of wave array
    
    vis2_FT_P1 = []
    vis2err_FT_P1 = []
    vis2_FT_P2 = []
    vis2err_FT_P2 = []    
    
    vis2_SC_P1 = []
    vis2err_SC_P1 = []
    vis2_SC_P2 = []
    vis2err_SC_P2 = []      
        
    ucoord = []
    vcoord = []
    
    flag_vis2_P1=[]
    flag_vis2_P2=[]

    for i in range(len(ff[8].data)):
        vis2_FT_P1.append(ff[8].data[i][4]) 
        vis2err_FT_P1.append(ff[8].data[i][5]) 
        vis2_FT_P2.append(ff[12].data[i][4]) 
        vis2err_FT_P2.append(ff[12].data[i][5])  

        vis2_SC_P1.append(ff[16].data[i][4]) 
        vis2err_SC_P1.append(ff[16].data[i][5]) 
        vis2_SC_P2.append(ff[20].data[i][4]) 
        vis2err_SC_P2.append(ff[20].data[i][5])  
        
        ucoord.append(ff[8].data[i][6])
        vcoord.append(ff[8].data[i][7])
        print(ff[8].data[i][6],ff[8].data[i][7])
        print(ff[12].data[i][6],ff[12].data[i][7])        


        
        flag_vis2_P1.append(ff[16].data[i][9])
        flag_vis2_P2.append(ff[20].data[i][9])

# flux for the FT and SC -> for P1 and P2 at 4 telescopes    
# each of variables is a list with 4-telescope datasets: each dataset is an array with 
# length of wave array

    flux_FT_P1 = []
    fluxerr_FT_P1 = []
    flux_FT_P2 = []
    fluxerr_FT_P2 = []  

    flux_SC_P1 = []
    fluxerr_SC_P1 = []
    flux_SC_P2 = []
    fluxerr_SC_P2 = []      
    
    for i in range(len(ff[10].data)):
        flux_FT_P1.append(ff[10].data[i][4])
        fluxerr_FT_P1.append(ff[10].data[i][5])
        flux_FT_P2.append(ff[14].data[i][4])
        fluxerr_FT_P2.append(ff[14].data[i][5])    

        flux_SC_P1.append(ff[18].data[i][4])
        fluxerr_SC_P1.append(ff[18].data[i][5])
        flux_SC_P2.append(ff[22].data[i][4])
        fluxerr_SC_P2.append(ff[22].data[i][5])  

# closure phase for the FT and SC -> for P1 and P2 at 4 telescopes    
# each of variables is a list with 4-telescope datasets: each dataset is an array with 
# length of wave array

    T3_FT_P1 = []
    T3err_FT_P1 = []
    T3_FT_P2 = []
    T3err_FT_P2 = []  

    T3_SC_P1 = []
    T3err_SC_P1 = []
    T3_SC_P2 = []
    T3err_SC_P2 = []      
    
    for i in range(len(ff[9].data)):
        T3_FT_P1.append(ff[9].data[i][6])
        T3err_FT_P1.append(ff[9].data[i][7])
        T3_FT_P2.append(ff[13].data[i][6])
        T3err_FT_P2.append(ff[13].data[i][7])    

        T3_SC_P1.append(ff[17].data[i][6])
        T3err_SC_P1.append(ff[17].data[i][7])
        T3_SC_P2.append(ff[21].data[i][6])
        T3err_SC_P2.append(ff[21].data[i][7])  

    
    return OB,wave_SC_P1,wave_SC_P2,wave_FT_P1,wave_FT_P2,vis2_FT_P1,vis2err_FT_P1,vis2_FT_P2,vis2err_FT_P2,vis2_SC_P1,vis2err_SC_P1,vis2_SC_P2,vis2err_SC_P2,ucoord,vcoord,flux_FT_P1,fluxerr_FT_P1,flux_FT_P2,fluxerr_FT_P2,flux_SC_P1,fluxerr_SC_P1,flux_SC_P2,fluxerr_SC_P2,T3_FT_P1,T3err_FT_P1,T3_FT_P2,T3err_FT_P2,T3_SC_P1,T3err_SC_P1,T3_SC_P2,T3err_SC_P2,flag_vis2_P1,flag_vis2_P2     
```

```python
#function taken from Kateryna, it calibrates the GRAVITY data

def calibration(l1,f1,l2,f2,alpha):

# l1,f1 - GRAVITY spectrum (in AA) -> shifted by RV star    
# l2,f2 - model spectrum (in AA)
# alpha and beta are np.arrays of coefficients of the function: alpha * lambda + beta

    chi=[]
    fin_alpha=[]
    fin_beta=[]
    corr=[]

    for k in range(1):#(len(alpha)):
#        a = abs(average(l1)*alpha[k]-average(l1)) # major offset, defined for each alpha
#        if alpha[k]>1.:
#            a = a*-1.
#        else:
#            a = a   
        alpha=1.
        beta = np.arange(-20.,+20.,0.5)
        for i in range(len(beta)):
            AMBER_l1 = alpha*l1+beta[i]   # define function for which coefficients will be searched
            
            # searching for overlapping regions 
            if min(l2)-min(AMBER_l1) <= 0.:
                new_AMBER_lam=[]
                new_AMBER_flux=[] 
                for j,jj in enumerate(AMBER_l1):
                    if jj<=max(l2):
                        new_AMBER_lam.append(jj)
                        new_AMBER_flux.append(f1[j])
            else:
                new_AMBER_lam=[]
                new_AMBER_flux=[] 

                for j,jj in enumerate(AMBER_l1):
                    if jj>=min(l2):
                        new_AMBER_lam.append(jj)
                        new_AMBER_flux.append(f1[j])    

            new_MARCS_flux=np.interp(new_AMBER_lam,l2,f2) # interpolate model np.array to new shifted observed grid
            
            #print beta[i]
            
            
            # METOD 1: cross-correlation of overlapping regions
            ycorr = np.correlate(new_AMBER_flux,new_MARCS_flux)
            corr.append(ycorr)
            
            # METOD 2: compute chi-square of overlapping regions
            chi.append(sum((new_AMBER_flux-new_MARCS_flux)**2))

            # save all alpha and beta
#            fin_alpha.append(alpha)
            fin_beta.append(beta[i])


    corr=np.array(corr)
    chi=np.array(chi)
    fin_beta=np.array(fin_beta)
#    fin_alpha=np.array(fin_alpha)
    
    # finds the best alpha and beta with the lowest chi2 or largest correlation 
#    alpha_chi2 = fin_alpha[chi.argmin()] 
#    alpha_corr = fin_alpha[corr.argmax()]
    beta_chi2 = fin_beta[chi.argmin()] 
    beta_corr = fin_beta[corr.argmax()] 

    #print beta_chi2,beta_corr


##    plt.figure()
##    plt.title('Wavelength calibration',fontsize=16)
##    
##    plt.plot(l1,np.interp(l1,l2,f2),color='g',label='MARCS ORIG',linestyle='--')
##    
##    plt.plot(l1-beta_corr,np.interp(l1,l2,f2),color='k',label='MARCS')
##
##    plt.plot(l1,f1,color='r',label='GRAVITY')
##
##    plt.xlabel('wavelength, A',fontsize=16)
##    plt.ylabel('Flux',fontsize=16)
##    plt.legend()

#    plt.figure()
#    plt.title('Wavelength calibration',fontsize=16)
#
#    plt.plot(l1,np.interp(l1,l2,f2),color='g',label='MARCS ORIG',linestyle='--')
#    
#    plt.plot(l1-beta_chi2,np.interp(l1,l2,f2),color='k',label='MARCS')
#
#    plt.plot(l1,f1,color='r',label='GRAVITY chi2')
#
#    plt.xlabel('wavelength, A',fontsize=16)
#    plt.ylabel('Flux',fontsize=16)
#    plt.legend()
    return beta_chi2,beta_corr
```

```python
######################################################################################
################################### FLUX CALIBRATION #################################
######################################################################################

#################################
# 2. Find shift between spectra and model tellurics for CAL1, CAL2 and SCI for all polarizations and all telescopes
#################################

# compute shifts using "calibrate" function
if find_shift == True:
    shift = []
    for i in range(len(flux_SC_P1)):
        a,b=calibration(np.array(wave_SC_P1),flux_SC_P1[i],np.array(wave_SC_P1),TEL_transm_interpol*max(flux_SC_P1[i]),1)
        #print('CAL1, telescope %s, P1')%((i+1),a)
        shift.append(a)
        a,b=calibration(np.array(wave_SC_P2),flux_SC_P2[i],np.array(wave_SC_P2),TEL_transm_interpol*max(flux_SC_P2[i]),1)
        #print('CAL1, telescope %s, P2')%((i+1),a)
        shift.append(a)
    #    if not CAL2_file == '': 
        a2,b2=calibration(np.array(wave_SC_P1_2),flux_SC_P1_2[i],np.array(wave_SC_P1_2),TEL_transm_interpol*max(flux_SC_P1_2[i]),1)
           # print('CAL2, telescope %s, P1')%((i+1),a2)
        shift.append(a2)
        a2,b2=calibration(np.array(wave_SC_P2_2),flux_SC_P2_2[i],np.array(wave_SC_P2_2),TEL_transm_interpol*max(flux_SC_P2_2[i]),1)
           # print('CAL2, telescope %s, P2')%((i+1),a2)
        shift.append(a2)
    
        a3,b3=calibration(np.array(wave_SC_P1_3),flux_SC_P1_3[i],np.array(wave_SC_P1_3),TEL_transm_interpol*max(flux_SC_P1_3[i]),1)
       # print('SCI, telescope %s, P1')%((i+1),a3)
        shift.append(a3)
        a3,b3=calibration(np.array(wave_SC_P2_3),flux_SC_P2_3[i],np.array(wave_SC_P2_3),TEL_transm_interpol*max(flux_SC_P2_3[i]),1)
       # print('SCI, telescope %s, P2')%((i+1),a3)    
        shift.append(a3)
    print(np.mean(shift))  
    shift = np.mean(shift)  



# shift all observations by a value derived above during telluric fit
wave_SC_P1 = np.array(wave_SC_P1) + shift
wave_SC_P1_3 = np.array(wave_SC_P1_3) + shift
wave_SC_P2 = np.array(wave_SC_P2) + shift
wave_SC_P2_3 = np.array(wave_SC_P2_3) + shift 
#if not CAL2_file == '': 
wave_SC_P1_2 = np.array(wave_SC_P1_2) + shift
wave_SC_P2_2 = np.array(wave_SC_P2_2) + shift

# interpolate MARCS spectrum to the wavelength scale of observations
CAL_MARCS_flux_interpol = np.interp(wave_SC_P1,CAL_MARCS_lam,CAL_MARCS_flux)
#if not CAL2_file == '': 
CAL2_MARCS_flux_interpol = np.interp(wave_SC_P1_2,CAL2_MARCS_lam,CAL2_MARCS_flux)    
##SCI_MARCS_flux_interpol = np.interp(wave_SC_P1_3,SCI_MARCS_lam,SCI_MARCS_flux) 
  



```

```python
#################################
# 3. Overplot fluxes for CAL1, CAL2 and SCI for all polarizations and all telescopes
#################################

if plot_all == True:
    plt.figure()
    for i in range(len(flux_SC_P1)):
        plt.plot(wave_SC_P1,flux_SC_P1[i])
        plt.plot(wave_SC_P2,flux_SC_P2[i],linestyle='--')
        
       # if not CAL2_file == '': 
        plt.plot(wave_SC_P1_2,flux_SC_P1_2[i])
        plt.plot(wave_SC_P2_2,flux_SC_P2_2[i],linestyle='--')

        plt.plot(wave_SC_P1_3,flux_SC_P1_3[i])
        plt.plot(wave_SC_P2_3,flux_SC_P2_3[i],linestyle='--')
        
    plt.plot(TEL_lam,TEL_transm*max(flux_SC_P1[i]),color='k',label='TELL')    
    plt.plot(wave_SC_P1,CAL_MARCS_flux_interpol/1.5,color='k',linewidth=2)
   # if not CAL2_file == '':
    plt.plot(wave_SC_P1,CAL2_MARCS_flux_interpol*1.5,color='r',linewidth=2)    
    plt.title('spectra')    

if plot_all == True:
    plt.figure()
    for i in range(len(vis2_SC_P1)):
        plt.plot(wave_SC_P1,vis2_SC_P1[i])
        plt.plot(wave_SC_P2,vis2_SC_P2[i],linestyle='--')
        
       # if not CAL2_file == '': 
        plt.plot(wave_SC_P1_2,vis2_SC_P1_2[i])
        plt.plot(wave_SC_P2_2,vis2_SC_P2_2[i],linestyle='--')

        plt.plot(wave_SC_P1_3,vis2_SC_P1_3[i])
        plt.plot(wave_SC_P2_3,vis2_SC_P2_3[i],linestyle='--')
    plt.title('squared visibilities')         
 
   

            
#################################
# 4. Compute TF for CAL2 and CAL1 for all polarizations and all telescopes
#################################

# *** a) CAL1 ***
# compute TF for all polarizations and all telescopes
plt.figure()  
TF_P1 = []
TF_P2 = []
for i in range(len(flux_SC_P1)):
    TF_P1.append(flux_SC_P1[i]/CAL_MARCS_flux_interpol)
    TF_P2.append(flux_SC_P2[i]/CAL_MARCS_flux_interpol)

    plt.plot(wave_SC_P1,flux_SC_P1[i]/CAL_MARCS_flux_interpol,label='TF CAL1 P1')
    plt.plot(wave_SC_P1,flux_SC_P2[i]/CAL_MARCS_flux_interpol,label='TF CAL1 P2',linestyle='--')
plt.xlabel('wavelength, A')
plt.ylabel('Flux')    

# compare to telluric spectrum
plt.plot(TEL_lam,TEL_transm*max(TF_P1[0]),color='k',label='TELL')    
plt.legend(loc=1)

# *** b) CAL2 ***
# compute TF for all polarizations and all telescopes 
if not CAL2_file == '': 
    plt.figure() 
    TF_P1_2 = []
    TF_P2_2 = []
    for i in range(len(flux_SC_P1_2)):
        TF_P1_2.append(flux_SC_P1_2[i]/CAL2_MARCS_flux_interpol)
        TF_P2_2.append(flux_SC_P2_2[i]/CAL2_MARCS_flux_interpol)
        
        plt.plot(wave_SC_P1_2,flux_SC_P1_2[i]/CAL2_MARCS_flux_interpol,label='TF CAL2 P1')
        plt.plot(wave_SC_P1_2,flux_SC_P2_2[i]/CAL2_MARCS_flux_interpol,label='TF CAL2 P2',linestyle='--')
    plt.xlabel('wavelength, A')
    plt.ylabel('Flux') 
    
    # compare to telluric spectrum
plt.plot(TEL_lam,TEL_transm*max(TF_P1_2[0]),color='k',label='TELL')    
plt.legend(loc=1)
# in the case if there is only one calibrator
#else:
#    TF_P1_2 = []
#    TF_P2_2 = []
#    fluxerr_SC_P1_2 = []
#    fluxerr_SC_P2_2 = []
#    CAL2_fluxerr = []
#    for i in range(len(flux_SC_P1)):
#        TF_P1_2.append(np.zeros(len(flux_SC_P1[i])))
#        TF_P2_2.append(np.zeros(len(flux_SC_P1[i])))
#        fluxerr_SC_P1_2.append(np.zeros(len(flux_SC_P1[i])))
#        fluxerr_SC_P2_2.append(np.zeros(len(flux_SC_P1[i])))

#################################
# 5. Calibrate CAL1 using TF of CAL2 and compare to MARCS of CAL1  -> comment in the case of only one calibrator
#################################

# a) NO SMOOTHING OF TF
if not CAL2_file == '':
    plt.figure()
    av = 0. # calibrated flux
    av_error = 0. # calibrated flux error
    for i in range(len(flux_SC_P1)):
        flux_SC_P1_calib = flux_SC_P1[i]/TF_P1_2[i]   
        plt.errorbar(wave_SC_P1,flux_SC_P1_calib,yerr=np.sqrt((fluxerr_SC_P1[i]/flux_SC_P1[i])**2+(fluxerr_SC_P1_2[i]/TF_P1_2[i])**2))
        av = av + np.array(flux_SC_P1_calib)
        # for each telescope and each polarization we compute error on calibrated flux
        # flux_cal = flux/TF => flux_cal_error = np.sqrt((flux_err/flux)**2+(TF_err/TF)**2) 
        # where TF_err = flux error for a calibrator used for TF computation
        av_error = av_error + ((fluxerr_SC_P1[i]/flux_SC_P1[i])**2+(fluxerr_SC_P1_2[i]/TF_P1_2[i])**2)
    
        flux_SC_P2_calib = flux_SC_P2[i]/TF_P2_2[i]   
        plt.errorbar(wave_SC_P1,flux_SC_P2_calib,yerr=np.sqrt((fluxerr_SC_P2[i]/flux_SC_P2[i])**2+(fluxerr_SC_P2_2[i]/TF_P2_2[i])**2),linestyle='--')
        av = av + np.array(flux_SC_P2_calib)
        av_error = av_error + ((fluxerr_SC_P2[i]/flux_SC_P2[i])**2+(fluxerr_SC_P2_2[i]/TF_P2_2[i])**2)

    flux_SC_P1_calib = av/(len(flux_SC_P1)*2) # calibrated flux
    flux_SC_P1_calib_error = np.sqrt(av_error)/(len(flux_SC_P1)*2)
   
    plt.errorbar(wave_SC_P1,flux_SC_P1_calib,yerr = flux_SC_P1_calib_error,color='darkblue',linewidth=2,label='AVERAGED CAL1')
    plt.plot(wave_SC_P1,CAL_MARCS_flux_interpol/max(CAL_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib),color='k',linewidth=2,label='MARCS CAL1')
       
    plt.title('calibration of CAL1 using TF of CAL2')
    plt.xlabel('wavelength, A')
    plt.ylabel('Flux') 
    plt.legend(loc=1)

#################################
# 6. Calibrate CAL2 using TF of CAL1 and compare to MARCS of CAL2
#################################

# a) NO SMOOTHING OF TF

    plt.figure()
    av2 = 0.
    av2_error = 0.
    for i in range(len(flux_SC_P1_2)):
        flux_SC_P1_2_calib = flux_SC_P1_2[i]/TF_P1[i]   
        plt.errorbar(wave_SC_P1_2,flux_SC_P1_2_calib,yerr=np.sqrt((fluxerr_SC_P1_2[i]/flux_SC_P1_2[i])**2+(fluxerr_SC_P1[i]/TF_P1[i])**2))
        av2 = av2 + np.array(flux_SC_P1_2_calib)
        av2_error = av2_error + ((fluxerr_SC_P1_2[i]/flux_SC_P1_2[i])**2+(fluxerr_SC_P1[i]/TF_P1[i])**2)
    
        flux_SC_P2_2_calib = flux_SC_P2_2[i]/TF_P2[i]   
        plt.errorbar(wave_SC_P1_2,flux_SC_P2_2_calib,yerr=np.sqrt((fluxerr_SC_P2_2[i]/flux_SC_P2_2[i])**2+(fluxerr_SC_P2[i]/TF_P2[i])**2),linestyle='--') 
        av2 = av2 + np.array(flux_SC_P2_2_calib)
        av2_error = av2_error + ((fluxerr_SC_P2_2[i]/flux_SC_P2_2[i])**2+(fluxerr_SC_P2[i]/TF_P2[i])**2)

    flux_SC_P1_calib_2 = av2/(len(flux_SC_P1_2)*2) # calibrated flux
    flux_SC_P1_calib_error_2 = np.sqrt(av2_error)/(len(flux_SC_P1_2)*2)
  
    plt.errorbar(wave_SC_P1_2,flux_SC_P1_calib_2,yerr=flux_SC_P1_calib_error_2,color='darkblue',linewidth=2,label='AVERAGED CAL2')  
    plt.plot(wave_SC_P1_2,CAL2_MARCS_flux_interpol/max(CAL2_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib_2),color='k',linewidth=2,label='MARCS CAL2')
    plt.title('calibration of CAL2 using TF of CAL1')
    plt.xlabel('wavelength, A')
    plt.ylabel('Flux') 
    plt.legend(loc=1)

#################################
# 7. Calibrate SCI using average TF of CAL1 and CAL2 and compare to MARCS of SCI
#################################

plt.figure()
av3 = 0.
av3_error = 0.
for i in range(len(flux_SC_P1_3)):
    flux_SC_P1_3_calib = flux_SC_P1_3[i]/((TF_P1[i]+TF_P1_2[i])/2.)  
    plt.errorbar(wave_SC_P1_3,flux_SC_P1_3_calib,yerr=np.sqrt((fluxerr_SC_P1_3[i]/flux_SC_P1_3[i])**2+(np.sqrt(fluxerr_SC_P1[i]**2+fluxerr_SC_P1_2[i]**2)/(TF_P1[i]+TF_P1_2[i]))**2))
    av3 = av3 + np.array(flux_SC_P1_3_calib)
    av3_error = av3_error + ((fluxerr_SC_P1_3[i]/flux_SC_P1_3[i])**2+(np.sqrt(fluxerr_SC_P1[i]**2+fluxerr_SC_P1_2[i])/(TF_P1[i]+TF_P1_2[i]))**2)
    
    flux_SC_P2_3_calib = flux_SC_P2_3[i]/((TF_P2[i]+TF_P2_2[i])/2.)   
    plt.errorbar(wave_SC_P1_3,flux_SC_P2_3_calib,yerr=np.sqrt((fluxerr_SC_P2_3[i]/flux_SC_P2_3[i])**2+(np.sqrt(fluxerr_SC_P2[i]**2+fluxerr_SC_P2_2[i])/(TF_P2[i]+TF_P2_2[i]))**2),linestyle='--') 
    av3 = av3 + np.array(flux_SC_P2_3_calib)
    av3_error = av3_error + ((fluxerr_SC_P2_3[i]/flux_SC_P2_3[i])**2+(np.sqrt(fluxerr_SC_P2[i]**2+fluxerr_SC_P2_2[i])/(TF_P2[i]+TF_P2_2[i]))**2)

flux_SC_P1_calib_3 = av3/(len(flux_SC_P1_3)*2) # calibrated flux
flux_SC_P1_calib_error_3 = np.sqrt(av3_error)/(len(flux_SC_P1_3)*2)

plt.errorbar(wave_SC_P1_3,flux_SC_P1_calib_3,yerr=flux_SC_P1_calib_error_3,color='darkblue',linewidth=2,label='AVERAGED SCI')       
##plt.plot(wave_SC_P1_3,SCI_MARCS_flux_interpol/max(SCI_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib_3),color='k',linewidth=2,label='MARCS SCI')
plt.title('calibration of SCI using average TF of CAL1 and CAL2')
plt.xlabel('wavelength, A')
plt.ylabel('Flux') 
plt.legend(loc=1)



###############################################################
############### PLOT RESULT OF FLUX CALIBRATION ###############
###############################################################
fig = plt.figure(figsize=(8,7))
gs = gridspec.GridSpec(3, 1, width_ratios=[1], height_ratios=[0.33,0.33,0.34],hspace=0.05) 

if not CAL2_file == '':
    ax1=plt.subplot(gs[0])

    ax1.errorbar(wave_SC_P1,flux_SC_P1_calib,yerr=flux_SC_P1_calib_error,color='darkred',linewidth=2,label='CAL1')  
    ax1.plot(wave_SC_P1,CAL_MARCS_flux_interpol/max(CAL_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib),color='k',linewidth=2,label='MARCS')
    ax1.set_ylabel('Flux')
    ax1.legend()

    ax2=plt.subplot(gs[1])
  
    ax2.errorbar(wave_SC_P1_2,flux_SC_P1_calib_2,yerr=flux_SC_P1_calib_error_2,color='darkred',linewidth=2,label='CAL2')  
    ax2.plot(wave_SC_P1_2,CAL2_MARCS_flux_interpol/max(CAL2_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib_2),color='k',linewidth=2,label='MARCS')
    ax2.set_ylabel('Flux')
    ax2.legend(loc=1)

ax3=plt.subplot(gs[2])

#OB_1,wave_SC_P1_1,wave_SC_P2_1,wave_FT_P1_1,wave_FT_P2_1,vis2_FT_P1_1,vis2err_FT_P1_1,vis2_FT_P2_1,vis2err_FT_P2_1,vis2_SC_P1_1,vis2err_SC_P1_1,vis2_SC_P2_1,vis2err_SC_P2_1,ucoord_1,vcoord_1,flux_FT_P1_1,fluxerr_FT_P1_1,flux_FT_P2_1,fluxerr_FT_P2_1,flux_SC_P1_1,fluxerr_SC_P1_1,flux_SC_P2_1,fluxerr_SC_P2_1,T3_FT_P1_1,T3err_FT_P1_1,T3_FT_P2_1,T3err_FT_P2_1,T3_SC_P1_1,T3err_SC_P1_1,T3_SC_P2_1,T3err_SC_P2_1,flag_vis2_P1_1,flag_vis2_P2_1 = read_FITS('SCI_SOri_SINGLE_SCI_VIS_CALIBRATED.fits')


ax3.errorbar(wave_SC_P1_3,flux_SC_P1_calib_3,yerr=flux_SC_P1_calib_error_3,color='darkred',linewidth=2,label='SCI') 
##ax3.plot(wave_SC_P1_3,SCI_MARCS_flux_interpol/max(SCI_MARCS_flux_interpol)*np.mean(flux_SC_P1_calib_3),color='k',linewidth=2,label='MARCS')
#ax3.plot(np.array(wave_SC_P1_1)+shift,(np.array(flux_SC_P1_1[0])+np.array(flux_SC_P2_1[0]))/2.*CAL_MARCS_flux_interpol)

#l,s,v,err,sys,lam=loadtxt('../Desktop/S_Ori_4_masks/AMBER_new4masks_mask2.dat',unpack=True,usecols=[0,1,2,3,4,5])
#for i,ii in enumerate(l):
#    ax3.axvline(ii,color='g')

ax3.set_xlabel(r'Wavelength [$\rm \AA$]')
ax3.set_ylabel('Flux')
ax3.legend(loc=1)


plt.show()
```
